a. Load the dataset into a Pandas DataFrame.
  
   Diabetes dataset taken from Professor's github dataset location

```python
import pandas as pd

# raw URL of  dataset diabetes.csv file from GitHub
github_url = 'https://raw.githubusercontent.com/Ishaanidisha/CS460-Machine-Learning/refs/heads/main/Datasets/diabetes.csv'

try:
    df = pd.read_csv(github_url)
    print("Dataset loaded successfully from GitHub.")
except Exception as e:
    print(f"Error loading dataset from GitHub: {e}")
```


b. Display the first 10 rows of the dataset.

```python
# Display first 10 rows of the dataset
print(df.head(10).to_markdown(index=False, numalign="left", stralign="left"))
```



c. Display the names of the columns in the dataset.

```python
# Display the column names of the DataFrame
print(df.columns.to_list())
```



d. Identify and print the target variable in the dataset.
Answer: The target variable is binary "Outcome"

```python
# Identify the target variable
target_variable = 'Outcome'

# Print the target variable
print(f"The target variable in the dataset is: {target_variable}")
```



e. Generate summary statistics for the dataset using the describe() method.

Answer:
The describe() method in Pandas is used to generate descriptive statistics of a DataFrame or Series. It provides a quick overview of the central tendency, dispersion, and shape of the distribution of a dataset's numerical columns.

For numerical data, the output of describe() includes:

*   count: The number of non-null observations
*   mean: The average value.
*   std: The standard deviation, which measures the amount of variation or dispersion.
*   min: The minimum value.
*   25% (Q1): The 25th percentile (first quartile).
*   50% (Q2): The 50th percentile (median).
*   75% (Q3): The 75th percentile (third quartile).
*   max: The maximum value.

This method is very useful for initial data exploration and for understanding the basic characteristics of your data.

```python
# Using describe method, generate summary statistics for the dataset
print(df.describe().to_markdown(numalign="left", stralign="left"))
```




f. Count and display the number of instances of each label in the target variable to check whether the data is balanced.

```python
# Count and display the number of instances of each label in the target variable
print(df[target_variable].value_counts().to_markdown(numalign="left", stralign="left"))
```



g. Plot a histogram for the target variable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style("whitegrid")

# Create a countplot for the 'Outcome' variable
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df, hue='Outcome', palette='viridis', legend=False)
# Add title and labels
plt.title('Distribution of Target Variable (Outcome)')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No Diabetes (0)', 'Diabetes (1)'])

# Save the plot
plt.savefig('outcome_distribution_countplot.png')
plt.show()
```



h. Check for any missing values in the dataset and provide the count for each column.

Answer: There are no missing values!


```python
# Check for missing values and sum them up for each column
missing_values = df.isnull().sum()

# Print the count of missing values for each column
print("Missing values in each column:\n")
print(missing_values.to_markdown(numalign="left", stralign="left"))
```


i. Calculate and print the mean, median, and standard deviation for at least three numerical columns.

Selected three columns : Glucose, Bloodpressure and Age



Additionally: Compute mean, median and standard deviation for all numerical columns.

```python
# Select three numerical columns
columns_to_analyze = ['Glucose', 'BloodPressure', 'Age']

# Calculate mean, median, and standard deviation for each selected column
mean_values = df[columns_to_analyze].mean()
median_values = df[columns_to_analyze].median()
std_dev_values = df[columns_to_analyze].std()

# Create a DataFrame to display the results
summary_df = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Standard Deviation': std_dev_values
})

# Print the results
print("Summary statistics for selected numerical columns:\n")
print(summary_df.to_markdown(numalign="left", stralign="left"))


#Additional computation for all numerical columns
print("\n\nAdditional computation for all numerical columns:\n")
# Calculate mean for all numerical columns
all_mean_values = df.mean()

# Calculate median for all numerical columns
all_median_values = df.median()

# Calculate standard deviation for all numerical columns
all_std_dev_values = df.std()

# Create a DataFrame to display the results for all columns
all_summary_df = pd.DataFrame({
    'Mean': all_mean_values,
    'Median': all_median_values,
    'Standard Deviation': all_std_dev_values
})

# Print the results
print("Summary statistics for all numerical columns:\n")
print(all_summary_df.to_markdown(numalign="left", stralign="left"))

```



j. Plot a simple line graph showing one feature (e.g., 'Glucose') over the first 50 samples.

```python
import matplotlib.pyplot as plt

# Select the 'Glucose' feature for the first 50 samples
glucose_first_50 = df['Glucose'].head(50)

# Create a simple line graph
plt.figure(figsize=(10, 6))
plt.plot(glucose_first_50.index, glucose_first_50.values, marker='o', linestyle='-')

# Add title and labels
plt.title('Glucose Levels Over First 50 Samples')
plt.xlabel('Sample Index')
plt.ylabel('Glucose Level')
plt.grid(True)

# Save the plot
plt.savefig('glucose_line_plot_first_50_samples.png')
plt.show()
```
